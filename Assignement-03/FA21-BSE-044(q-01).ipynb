```python
# 4/17/2024
# CSC354 – Assignmen3 – ML – Support Vector Machines
# YAHYA IRFAN 
# FA21-BSE-044
# Use the dataset (dataset-q-1.csv) and fit a linear SVM (using default parameter settings) on the entire training 
# data. (Note: You just have to fit (train) the model on the entire dataset, no evaluation here)
# 1. Do the positive and negative instances group together, suggesting a clear separation between the 
# classes?
# 2. Are there any outliers? If yes, can you spot them?
# Hint: Plotting the decision boundary and training data points using a scatter plot may help you answer the questions


import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
import numpy as np

data = pd.read_csv('dataset-q-1.csv', header=None)
X = data.iloc[:, :-1]
y = data.iloc[:, -1]

svm_model = SVC(kernel='linear')
svm_model.fit(X, y)


def plot_decision_boundary(X, y, model):
    x_min, x_max = X.iloc[:, 0].min() - 1, X.iloc[:, 0].max() + 1
    y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.8)

    plt.scatter(X[y == 1].iloc[:, 0], X[y == 1].iloc[:, 1], c='b', marker='o', label='Positive Class')
    plt.scatter(X[y == -1].iloc[:, 0], X[y == -1].iloc[:, 1], c='r', marker='x', label='Negative Class')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.title('Decision Boundary and Data Points')
    plt.show()

plot_decision_boundary(X, y, svm_model)
```
