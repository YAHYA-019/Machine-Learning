```python
# 4/17/2024
# CSC354  Assignmen3  ML  Support Vector Machines
# YAHYA IRFAN 
# FA21-BSE-044
# Rerun the experiment from Q1 with C = 0.01.
# 1. Plot again and note down your findings. Does the hyperplane misclassify (handle) any outliers?
# 2. Try increasing the value of C (100, 300, 700, 1000) and observe the effect on the resulting hyperplanes.
# Write down your findings


import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
import numpy as np

data = pd.read_csv('dataset-q-1.csv', header=None)
X = data.iloc[:, :-1]
y = data.iloc[:, -1]

# C = 0.01
svm_model_c_001 = SVC(kernel='linear', C=0.01)
svm_model_c_001.fit(X, y)


def plot_decision_boundary(X, y, model):
    x_min, x_max = X.iloc[:, 0].min() - 1, X.iloc[:, 0].max() + 1
    y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.8)

    plt.scatter(X[y == 1].iloc[:, 0], X[y == 1].iloc[:, 1], c='b', marker='o', label='Positive Class')
    plt.scatter(X[y == -1].iloc[:, 0], X[y == -1].iloc[:, 1], c='r', marker='x', label='Negative Class')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.title('Decision Boundary and Data Points')
    plt.show()


# C = 0.01
plot_decision_boundary(X, y, svm_model_c_001)


C_values = [100, 300, 700, 1000]

for C_value in C_values:
    svm_model_c = SVC(kernel='linear', C=C_value)
    svm_model_c.fit(X, y)
    plot_decision_boundary(X, y, svm_model_c)



```

Answers:
1. When C = 0.01:
When C is very small (e.g., C = 0.01), the margin is very wide, and the classifier gives more chances of misclassifications. As in the plotting there is an outlier where a positive class data point comes to negative class region.

2. When C is (100, 300, 700, 1000):
As C increases, the margin becomes narrower, and the classifier gives less chances of  of misclassifications.
As we can see in all the plots there isn,t any outlier present.
